# K2 lightcurves

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl
import matplotlib
matplotlib.rcParams['font.size'] = 12

import chronos as cr
```

## sap & pdcsap

```python
epicid=247047380
campaign=13
star = cr.K2(epicid=epicid, campaign=campaign)
lc = star.get_lc(lctype="pdcsap")
lc.scatter()
```

## everest

```python
eve = cr.Everest(epicid=epicid, campaign=campaign, verbose=False)
lc = eve.lc_everest
lc.scatter()
```

```python
eve.everest_recarray
```

## K2SFF

```python
k2sff = cr.K2sff(epicid=epicid, campaign=campaign, verbose=False)
lc = k2sff.lc_k2sff
lc.scatter()
```

```python
k2sff.k2sff_recarray
```
