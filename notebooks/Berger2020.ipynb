We identify 35 confirmed planets and 15 planet
candidates that occupy the hot sub-Neptunian
desert (Lundkvist et al. 2016). We determine that
most of the desert planets orbit evolved stars, and
thus it is unlikely that they are young planets currently undergoing mass loss. 

```python
from astropy.table import Table

tab = Table.read("../chronos/data/Berger2020/Berger2020.txt", format="ascii")
tab
```

The hot sub-Neptunian desert, defined as Rp = 2.2–
3.8 R⊕ and Fp > 650 F⊕, is another region of parameter
space believed to be devoid of planets (Lundkvist et al.
2016).

```python
df = tab.to_pandas()
```

```python
df['Flag'].unique()
```

```python
idx = (df['Flag']=='Young') | (df['Flag']=='HZYoung') | (df['Flag']=='YoungDesert')
df[idx]
```

```python
r = (df['Radius']>2.2) & (df['Radius']<3.8)
f = df['Flux']>650
idx = r & f
groups = df[idx].groupby('PD').groups
{g: len(groups[g]) for g in groups}
```

```python
df.iloc[groups['CONFIRMED']]
```
