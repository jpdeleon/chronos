```python
%matplotlib inline
import matplotlib.pyplot as pl
import numpy as np

import chronos as cr
```

# target pixel file (TPF)

```python
epicid = 211399359
t = cr.K2(epicid=epicid)
```

```python
tpf = t.get_tpf()
tpf.plot()
```

```python
t.tpf.plot(aperture_mask=t.tpf.pipeline_mask)
```

# Everest pipeline

```python
e_c5 = cr.Everest(epicid=epicid)
tpf_e_c5 = e_c5.get_tpf()
tpf_e_c5.plot()
```

```python
e_c18 = cr.Everest(epicid=epicid, campaign=18)
tpf_e_c18 = e_c18.get_tpf()
tpf_e_c18.plot()
```

## pipeline mask

Compare: https://archive.stsci.edu/k2/hlsp/everest/search.php

```python
fig, ax = pl.subplots(1, 2)

#wrong?
ax[0].matshow(tpf_e_c5.pipeline_mask, origin='lower', label='C5')
ax[1].matshow(tpf_e_c18.pipeline_mask, origin='lower', label='C16')
```

* c5: https://archive.stsci.edu/prepds/everest/v2/html/c05/ep211399359-lc.html
* c18: https://archive.stsci.edu/prepds/everest/v2/html/c18/ep211399359-lc.html

```python
fig, ax = pl.subplots(1, 2)

ax[0].matshow(e_c5.everest_best_aper_mask, origin='lower', label='C5')
ax[1].matshow(e_c18.everest_best_aper_mask, origin='lower', label='C16')
```

## gaia sources

```python
ax = e_c5.plot_gaia_sources_on_survey()
```

```python
ax = e_c18.plot_gaia_sources_on_survey()
```

# K2SFF pipeline

```python
k_c5 = cr.K2sff(epicid=epicid)
tpf_k_c5 = k_c5.get_tpf()
tpf_k_c5.plot()
```

```python
k_c18 = cr.K2sff(epicid=epicid, campaign=18)
tpf_k_c18 = k_c18.get_tpf()
tpf_k_c18.plot()
```

## pipeline mask

* c5: https://archive.stsci.edu/prepds/k2sff/html/c05/ep211399359.html
* c18: https://archive.stsci.edu/prepds/k2sff/html/c18/ep211399359.html

```python
fig, ax = pl.subplots(1, 2)

ax[0].matshow(k_c5.k2sff_best_aper_mask, origin='lower', label='C5')
ax[1].matshow(k_c18.k2sff_best_aper_mask, origin='lower', label='C16')
```

# gaia sources

```python
ax = k_c5.plot_gaia_sources_on_survey()
```

```python
ax = k_c18.plot_gaia_sources_on_survey()
```

# comparison

```python
fig, axs = pl.subplots(2, 2, sharex='row', sharey='col')
ax = axs.flatten()

ax[0].matshow(e_c5.everest_best_aper_mask, origin='lower', label='everest')
ax[1].matshow(k_c5.k2sff_best_aper_mask, origin='lower', label='k2sff')

ax[0].set_title('Everest')
ax[1].set_title('K2SFF')

ax[2].matshow(e_c18.everest_best_aper_mask, origin='lower', label='everest')
ax[3].matshow(k_c18.k2sff_best_aper_mask, origin='lower', label='k2sff')
```
